```python
df_carseat['Urban'] = df_carseat.Urban.map(dict(Yes=1, No=0))
df_carseat['US'] = df_carseat.US.map(dict(Yes=1, No=0))
```

Linear Regression Model might be a good choice. The equation is:

**sales = β0 + β1×Price + β2×Advertising + β3×Urban + β4×US**

```python
X_carseat = df_carseat[['Price', 'Advertising','Urban', 'US']]
y_carseat = df_carseat['Sales']
```

```python
import statsmodels.api as sm
from scipy import stats
```

```python
X_carseat = sm.add_constant(X_carseat)
model_carseat = sm.OLS(y_carseat,X_carseat)
res_carseat = model_carseat.fit()
```

```python
print(res_carseat.summary())
```

With the table above, the equation is:

**sales = 13.0113 - 0.0546×Price + 0.1203×Advertising  - 0.0388×Urban + 0.0585×US**

The coef of Advertising is 0.1203. It means that with the price increase for 1000 units, the sale will increase for about 120.3 units.

The coef of Urban is -0.0388. It means that a urban store will have about 0.0388 units less sale comparing with a rural one.

The coef of US is 0.0585. It means that a US store will have about 0.0585 units more sale comparing with one that is not in US.
